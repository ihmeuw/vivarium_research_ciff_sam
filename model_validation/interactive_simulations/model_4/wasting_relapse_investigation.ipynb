```python
from datetime import datetime, timedelta
from pathlib import Path

import itertools

from vivarium import Artifact, InteractiveContext
import ipywidgets
import pandas as pd, numpy as np
import seaborn as sns
pd.set_option('display.max_rows', 8)
```

```python
path = Path('/ihme/homes/alibow/vivarium_ciff_sam/src/vivarium_ciff_sam/model_specifications/ciff_sam.yaml')
path
```

# x-factor = 1.0

```python
age_cut=4-30/365
sim = InteractiveContext(path)
for i in list(range(1,730+30)):
    sim.step()
    pop = sim.get_population()
    
    # add who develop mam to cohort
        # currently unconditional on successful treatment
    if i == 1:
        cohort_start = (pop.loc[(pop['moderate_acute_malnutrition_event_count']==1)
                               & (pop['age'] < age_cut)]
                        ['moderate_acute_malnutrition_event_time'])
    else:
        cohort_start = cohort_start.append(pop[~pop.index.isin(cohort_start.index)]
                            .loc[(pop['moderate_acute_malnutrition_event_count']==1)
                                & (pop['age'] < age_cut)]
                                           ['moderate_acute_malnutrition_event_time'])
```

```python
eps_end = pd.concat([pop.loc[:, ['age', 'sex', 
                              'moderate_acute_malnutrition_event_count',
                              'moderate_acute_malnutrition_event_time',
                              'severe_acute_malnutrition_event_count',
                              'severe_acute_malnutrition_event_time',
                              'cause_of_death']], 
                 sim.get_value('x_factor.exposure')(pop.index),
                 sim.get_value('wasting_treatment.exposure')(pop.index)], axis=1)
eps_end
```

```python
cohort_end = (pd.DataFrame(cohort_start)
        .rename(columns={'moderate_acute_malnutrition_event_time':'first_episode_time'})
        .merge(eps_end, right_index=True, left_index=True))
cohort_end = cohort_end.loc[cohort_end.first_episode_time<datetime(year=2022,month=2,day=1)]
cohort_end['moderate_acute_malnutrition_event_count'] = cohort_end['moderate_acute_malnutrition_event_count'] - 1
cohort_end
```
